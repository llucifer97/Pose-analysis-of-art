```python
def delete_pose(k):

    path_to_json = './json_single_pose/'
    json_files = [pos_json for pos_json in os.listdir(path_to_json) if pos_json.endswith('.json')]
    json_arr2 = list()
    for json_file in json_files:
        json_file_path = os.path.join(path_to_json, json_file)
        with open(json_file_path, "r") as f:
            json_arr2.append(json.load(f))
    file_id = []
    for j in range(len(json_arr2)):
        file = natsorted(list(json_arr2[j].items()))[0][1]
        id = natsorted(list(json_arr2[j].items()))[1][1]
        file_id.append([file,id])
    ids = []
    for i in range(len(file_id)):
        ids.append(int(file_id[i][1]))
    if k in ids:
        idx = ids.index(k)
        file_list = os.listdir('./json_single_pose/')
        os.remove('./json_single_pose/' + file_list[idx] )
        print("Deleted!")
        print(j)
    else:
        print("File does not exit!")
```
